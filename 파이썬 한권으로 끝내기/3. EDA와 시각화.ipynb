```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
iris['class'] = load_iris().target

iris['class'] = iris['class'].map({0:'setora', 1:'virsicolor', 2:'virginica'})
iris

def linear_func(x) :
    return 2 * x + 1

X=iris['sepal length (cm)']
plt.plot(X,linear_func(X), c='#789395')
plt.show()
```

```python
import numpy as np
X,Y = iris['sepal length (cm)'], iris['petal length (cm)']
b1, b0 = np.polyfit(X,Y,1)
plt.scatter(x=X, y=Y, alpha=0.5)
plt.plot(X, b1*X+b0, color='red')
plt.show()
```

```python
iris2=iris.sort_values(by='sepal length (cm)')
X,Y = iris2['sepal length (cm)'], iris2['petal length (cm)']
b2, b1, b0 = np.polyfit(X,Y,2)
plt.scatter(x=X, y=Y, alpha=0.5)
plt.plot(X, b2*X*X + b1*X + b0, color='red')
plt.show()
```

```python
plt.plot('sepal length (cm)', 'petal width (cm)', data=iris2)
plt.show()
```

```python
plt.plot('sepal length (cm)', 'petal width (cm)', data=iris2.loc[iris2['class']=='setora'])
plt.plot('sepal length (cm)', 'petal width (cm)', data=iris2.loc[iris2['class']=='virsicolor'])
plt.plot('sepal length (cm)', 'petal width (cm)', data=iris2.loc[iris2['class']=='virginica'])
plt.show()
```
