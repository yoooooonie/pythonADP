## 1-1. 시각화 포함 탐색적 자료분석을 시행하시오. (EDA)

```python
import matplotlib.pyplot as plt

plt.bar(df['grade'].value_counts().index, df['grade'].value_counts().values)
plt.xlabel('grade')
plt.ylabel('student')
```

```python
import scipy.stats as stats
stats.shapiro(df['grade'])
```

```python
5000건 이하의 소규모 데이터의 정규성을 검정하기 위해 Shapiro 검정을 진행한다. 
p값이 0.001로 0.05 이하이기 때문에 해당 데이터는 정규성을 띄고 있음을 확인할 수 있다. 
```

### 결측치를 예측하는 방법
1. 단순 대치법
수치형 변수라면 각 컬럼의 평균이나 중앙값을 사용하여 결측치를 보관할 수 있으며, 명목형, 범주형 변수라면 최빈값을 사용하여 대치할 수 있다.
2. KNN을 이용한 결측치 대체
보간법 중 결측치가 없는 컬럼들의 최근접 이웃 알고리즘을 통해 결측치가 있는 변수를 대체할 수 있다. 단, KNN은 거리 계산이 가능한 수치형 변수인 경우만 사용이 가능하다.

모든 행이 결측치인 행은 없으니 결측치가 있는 컬럼에 데이터를 채우도록 한다. 수치형 변수에만 결측치가 있어 KNN을 사용한다.

```python
# !conda install scikit_learn
from sklearn.impute import KNNImputer

KNN_data = df.select_dtypes(exclude='object')
KNN_data

imputer = KNNImputer()
df_filled = imputer.fit_transform(KNN_data)
df_filled = pd.DataFrame(df_filled, columns=KNN_data.columns)
df[KNN_data.columns] = df_filled

df.isna().sum()
```
